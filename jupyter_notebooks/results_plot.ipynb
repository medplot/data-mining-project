```python
import pandas as pd
```

```python
results = [ ["K-Nearest Neighbors", 0.587],
            ["Nearest Centroid", 0.574],
            ["Decision Tree ", 0.594],
            ["Random Forest", 0.601],
            ["XGBoost", 0.608],
            ["Support Vector Classifier", 0]   ,
            ["Naive Bayes", 0],
            ["Neural network with embeddings", 0.865],
            ["Neural network one hot encoding", 0.824]]

df = pd.DataFrame(results, columns=["Model", "F2-Score"])
df.sort_values('F2-Score', ascending = True, inplace= True)
ax = df.plot.barh(x="Model", y="F2-Score", title="Model comparison based on the F2-Score")
ax.legend(loc="lower right")
for i, v in enumerate(df["F2-Score"]):
    ax.text(v, i, " "+str(v), va="center")
ax.set_xlim(0,1)

fig = ax.get_figure()
fig.savefig('../images/results_plot.png', dpi=300, bbox_inches="tight")
```
